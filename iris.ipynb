# **HISTOGRAM**



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
iris.keys()
data = iris.data
targets = iris.target
colors = ['red' , 'yellow' , 'blue']
for r in range(4):
    for x, color in zip(range(len(targets)), colors):
        plt.hist(data[targets == x, r], color = color)
        plt.xlabel(iris.feature_names[r])
        plt.legend(iris["target_names"])
    plt.show()
```

# **scatter plot every 2 attributes together**

```python
from itertools import combinations
att_comb=[0,1,2,3]
for comb in combinations(att_comb, 2):
     plt.scatter(x=data.T[comb[0]],y=data.T[comb[1]],c=iris.target)
     plt.xlabel(iris.feature_names[comb[0]])
     plt.ylabel(iris.feature_names[comb[1]])
     plt.show()
```

# **correlation matrix**

```python
import seaborn as sns
for label in range(len(iris.target_names)):
  corr= np.corrcoef(data[label == iris.target].T)
  print(corr)
sns.heatmap(corr)

#Positive Correlation: both variables change in the same direction.
#Neutral Correlation(0 or 1): No relationship in the change of the variables.
#Negative Correlation: variables change in opposite directions.
```
